## Betalist Webscraper

#### Import Libraries

```python
import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

#### The Field Types for my Webscraper 
> - `company name`: takes the company name from the markets list
> - `description`: describes the company from betalist
> - Betalist Link: adds a link to the original company posting on betalist
> - Crunchbase Link: adds a link to the company on crunchbase


```python
company_name = []
description = []
betalist_link = []
crunchbase_link = []
```

#### The URL we will be Scraping
> - `market`:the type of market that we are looking to scrape
> - `default`: the format of the web address for betalist
> - `urls`: the array including the subpages of the market

```python
market = "finance-technology"
default = "https://betalist.com/markets/"+market+"?page="
urls = [default]
```

```python
#adds subpages to an array  
for i in range(1,21):
  urls.append(default+str(i))

def companies(url):
  results = requests.get(url)
  soup = BeautifulSoup(results.text , "html.parser")
  company = soup.find_all('div', class_ ="startupCard")
  return company
```

#### The Components of the Company List
> - `def names`: function for scraping company names
> - `def des`: function for scraping company descriptions
> - `companies`: creates a dataframe to be used for the csv with the all information 


```python
#appends individual company names
def names(companies):
  for container in companies:
    name = container.find("a", class_="startupCard__details__name").text
    company_name.append(str.rstrip(name))
  return company_name

#appends individual company descriptions 
def des(companies):
  for container in companies:
    des = container.find("a", class_ ="startupCard__details__pitch").text
    description.append(des)
  return description

for i in range(len(urls)):
  page = companies(urls[i])
  company_name = names(page)
  description = des(page)
```

#### Runs a loop to create arrays for links 

```python
for i in range(len(company_name)):
    name_lower = company_name[i].lower()
    adjusted_name = name_lower.replace(" ", "-")
    
    betalink = "https://betalist.com/startups/"+adjusted_name
    cblink = "https://www.crunchbase.com/organization/"+adjusted_name
    
    betalist_link.append(str(betalink))
    crunchbase_link.append(str(cblink))
```



#### Creates dataframe and spreadsheet

```python
companies = pd.DataFrame({
'name': company_name,
'description': description,
'betalist_link': betalist_link,
'crunchbase_link': crunchbase_link,
})

companies.to_csv('company_list.csv')
companies
```
